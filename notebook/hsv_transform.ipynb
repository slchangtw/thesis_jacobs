```python
import cv2

from keras.preprocessing import image

import matplotlib.pyplot as plt
```

```python
def plot_hsv_transfrom(file):
    img = image.load_img(file)
    img_tensor = image.img_to_array(img) * 1./255
    
    img_new = cv2.cvtColor(img_tensor, cv2.COLOR_RGB2HSV)
    
    plt.imshow(img_new)
    plt.axis('off')
    plt.savefig('../data/output/result.jpg')
    
    return None
```

```python
plot_hsv_transfrom('../data/input/Training/Pear/0_100.jpg')
```

```python
plot_hsv_transfrom('../data/input/Training/Pear Abate/6_100.jpg')
```

```python
plot_hsv_transfrom('../data/input/Training/Pear Monster/0_100.jpg')
```
